###### We can't use date format(01-06-2013) in a Mathematical Model

To encode date first we have to convert it into a proper format

```python
data["Date"].dtypes
```

Here our date is in the object format

```python
pd.to_datetime(data["Date"]) # Column will be converted to date and time format
```

```python
data["Date"] = pd.to_datetime(data["Date"])
```

Now we are encoding it into format that a machine learning model understands

```python
data.dtypes
```

Now our date is in datetime format. Now we need to convert it into numerical format in which it can be used in a machine learning model

```python
data["Date"].dt.year # Extract only the year
```

#### $dt$ is Used to access the datetime properties

```python
data["Date"].dt.month
```

```python
data["Date"].dt.day
```

```python
data["Year"] = data["Date"].dt.year # adding the year column
```

```python
data["Month"] = data["Date"].dt.month
```

```python
data["Day"] = data["Date"].dt.day
```

```python
data
```

What all information we need from the date is properly extracted now we can delete the date column

```python
data.drop("Date",axis=1,inplace=True)
```
