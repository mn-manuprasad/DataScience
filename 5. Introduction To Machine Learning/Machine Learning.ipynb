## 1. Supervised Learning

When we are working with a labeled data that is known as Supervised learning 
Data which has both inputs(Variables which we used to estimate the target)  and output(Target)

Let's assume that you are trying to estimate the price of a house

Conditions (Parameters) we check while estimating the House Price 

    1. BHK(Bedroom,Hall,Kitchen)
    
    2. Location
    
    3. Area Of House
    
    4. Facilities  etc..
    
Using this parameters we want the machine to estimate(Predict) the price of a house.

Here 
    1. BHK(Bedroom,Hall,Kitchen)
    
    2. Location
    
    3. Area Of House
    
    4. Facilities  etc..
    
    Above are inputs and price is the output
    
When you are having the historical(Previous) Data (Both Inputs and Outputs) to make your machine learn, That is called supervised learning



                         Supervised Learning

## In Machine Learning ,

### $ Inputs \ are \ called \ Independent \ variable \ or \ Features $ 


### $ Outputs \ are \ called \ Dependent \ variable \ or \ target $

Why it is called $ Independent \ variable $ because all these variables should be independent 

Does BHK depends upon Location ?

Does BHK depends upon area ?

Does BHK depends upon facilities ?

So Inputs are called the independent variables because ideally all of them are independent to each other

###### Comes to Dependent variable,

Price Depends upon all the input variables . That is why it is called dependent variable. The target variable we want to predict which depends upon the independent(input) variables is called dependent variable

###  Supervised Learning is of two types

$1. Regression $ 

$2. Classification $

### Regression

######  If the target variable(Dependent) is numerical in nature it is regression analysis. Example House Price. If the dependent variable is numerical then it is regression

### Classification

###### If the target variable is categorical it is Classification. Example- Whether customer will purchase product or not. If the dependent variable is categorical then it is classification

# Reinforcement Learning

It is the real time learning. In the absence of a training dataset, it is bound to learn from its experience. Example: The problem is as follows: We have an agent and a reward, with many hurdles in between. The agent is supposed to find the best possible path to reach the reward.It is a reward based learning. Used in Game AI, Robot Navigation, Real Time Decisions etc
