## Mean

What exactly mean indicate 

Mean = Sum of all data points /  Number of data Points
 
What exactly mean represent ?

Mean is a data point which represents the entire data in a single data point(Which is also called a point estimate)

In majority of the cases, when the data is ideal , the mean will be the center of the entire data

mean comes under measure of central tendency

In general, Most of the data will be near to its mean

##### Outliers


Any data point which is far away from your regular data set is known as the outliers

Here one particular student salary is far away from the regular student salary. As it is far away from the regular salary, this particular student(student salary) is called an outliers

When a few data point(s) which is far away from most of the population is called an outliers

The disadvantage of mean is, it is affected by the outliers
Outliers can be either on left or right side

If the outliers is on the right side then the mean will be dragged to the right side and vise versa

###### To Avoid These problem we use median


###### In Median we don't consider all data points, In mean we consider all data points, So in that way the outliers effect will be avoided in the median

```python
np.median(sal)
```

Without affect of outliers we will be getting the exact center values

if the mean is higher or lower than the median then we can conclude that definitely there will be some outliers

In an Ideal Case the mean and median should be equal

#### General Steps in Calculating Median

1. Sort the data in ascending order
2. Find the center value of data 
3. If we have two center values then take the average of both

### Variance is the average deviation from each data point to mean

In Variance we use the mean as a reference point to understand how each data point is deviated from the center or how much is the average spread of data points from the center

#### The deviation formula is  :    $Xi -  \bar{X}$ (read us X-bar)
#### Here
#### $ Xi $ ----> Data Points
#### $\bar{X}$ ----> Mean of the data points

Deviation can be on any Direction
The below diagram shows the deviation of data points from center to both directions

Deviation means distance a negative or positive sign indicates the direction of deviation.The magnitude can't be negative 

Let's  assume that the data points are as follows

10,20,30,40,50

Let's assume Mean = 30 ( $\overline{X} $ )

So the deviation of every data point from mean is ($Xi -  \overline{X}$)

10 -->  10-30 = -20

20 -->  20-30 = -10

30 -->  30-30 =  0

40 -->  40-30 = 10

50 -->  50- 30 = 20


Here the deviation of data points 10 and 20 from mean is negative values

The deviation Can't be negative the sign shows the direction of deviation. So to remove the sign(To Only Take the Magnitude) We Square the value 

ie,

 $(Xi -  \bar{X})^2$

Variance is a point estimate . Here we have deviation of every point from the mean value. We square each deviation(To remove negative sign) and add then divide it by number of data points
That is



## $\frac{(10-30)^2 + (20-30)^2 + (30-30)^2 + (40-30)^2 + (50- 30)^2} {5(Number of data points)} $

#### $ Here \ the \ variance \ is \ \frac{1000}{5} = 200 $

#  $ Variance= \frac{\sum_{} {}(Xi -  \overline{X})^2}{n}$

###### Variance is the average deviation from each data point to the mean. Variance can be only calculated for the numerical data

### Variance is not the actual deviation but the squared deviation. To avoid this problem we have Standard Deviation

# Standard Deviation (SD)

In Standard Deviation , we get the actual deviation not the squared deviation. So SD is the average deviation of data points from mean value

# Population Variance and Sample Variance


## Population Variance

###### If we are taking the entire population to find variance then it is called population variance

##  $ \sigma^2 = \frac{\sum_{i=1}^{N}(x_{i}-\mu)^2}{N}$ 

$ \sigma^2 = Population \ Variance $ \
$ x_{i} = Value \ of \ i^{th} \ element$ \
$ \mu = Population \ Mean $ \
$ N = Population Size$

## Sample Variance

###### If we are only taking a sample from the entire population to find variance then it is called Sample variance

##  $ S^2 = \frac{\sum_{i=1}^{n}(x_{i}-\overline{x})^2}{n-1}$ 

$ S^2 = Sample \ Variance $ \
$ x_{i} = Value \ of \ i^{th} \ element$ \
$ \overline{x} = Population \ Mean $ \
$ n = Sample Size$

Why  $ (n-1) $ in sample variance ?

Based various researches and analysis it found that instead of $n$ if we are using $n-1$ there are chances that the sample variance will be almost equal to population variance

###### If we are talking about the entire data then we can use population ($N$) statistic. If We are only talking about a subset of the population(Sample) then we can use sample ($n$) statistic
