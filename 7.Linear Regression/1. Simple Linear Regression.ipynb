Here based on the years of experience of a person we have to estimate the salary

        YearsExperience -> Independent Variable
        
        Salary          -> Dependent Variable

General Equation of Simple Linear Regression have many notations,

### $ y = a_0 + a_1 * x $

### $ y = mx + c $

### $ y = b_0 + b_1*x $

### $ y = \theta_0 + \theta_1 * x $

    Whichever notation we take, The Concept is same


If we take $y = mx +c$. Here,
    
      1. m   -> Slope
      
      2. c   -> Constant
      
      3. y   -> Dependent Variable(Output)
      
      4. x   -> Independent Variable(Inputs)
      
    

In the above use case the equation can be written us,

$Salary = (m * Years \ Of \ Experience) + c$

Let's take an example if Experience=4 what all are things we need to know to estimate the salary

$Salary = (m * 4) + c$

We need to find the m and c to estimate the salary. So our goal is to identify the best m and c values for the data

## How to calculate the best m and c value  using Simple Linear Regression Algorithm?

We can apply Simple Linear Regression algorithm in 2 ways

     1. OLS (Ordinary Least Squre)
     
     2. Gradient Descent Approach

```python
sns.scatterplot(data["YearsExperience"],data["Salary"])
```

We plot a graph(Scatter Plot) with Salary in X-axis and YearsExperience in Y-axis.

## 1. At first we need to create a random line 

The + indicated in the graph is the actual data points(Salary)

## 2. Next step is to find out the deviation (distance) form the green points to the line

               Assume that the data points are called y and corresponding point on the line are called y'

    y is our actual value

    y' is the respective prediction point
    
    and (y-y') is the deviation(Distance) of data point from the line

Can the y-y' be negative , Distance can't be negative so apply square to it ie, $(y-y')^2$

In this way we find out the deviation of all points from the line

Here from the above graph we are getting seven deviations( Because there are 7 Data points). Assume that the D1,D2,D3,D4,D5,D6,D7 contains the respective deviation of each data points

### So the deviation of the entire line is = $ \frac{D1+D2+D3+D4+D5+D6+D7}{7} $ is called the $ Deviation \ of \ line ( \ Error \ or \ Loss \ or \ Cost function \ or \ Mean \ Squared \ Error)$

## $ Cost Function(Error,Loss)\ of \ line -> e = \frac{\sum_{}{} (y-y')^2} {N}$ 

      N -> No of Data points

Now we got the error of the line lets assume that the error is 50000 for the above line , Which means on average there is 50000  difference between the actual point(Actual Value) and line (Predicted Value). Now we try to shift the line to check whether there is any other line with less error(Best Fit Line).Like this we repeatedly shift the line and calculate the deviation of the line.So that we identify the error for all possible lines , last we will be having a list of errors for corresponding lines[e1,e2,e3,.....]. From which we select the line with minimum error. That line is called the $Best \ Fit \ Line$

## 3. Once we got the best fit line , the next immediate step is to calculate the m and c values

y = mx + c. In this equation, What exactly is c ?

if we  put the value of x as 0, then y=c. The c is called y intersect because it is a line where your point is intersecting y-axis when the x = 0

When Experience = 0, Then the salary will be C.

ie, The initial salary given to an employee with experience is c

So with Experience

Salary  = C + Salary_For_Added_Experience

m --> Slope 

m is like weight

How much weightage  the company is giving for the experience 

In general what exactly is slope ?


## $ Equation \ of \ slope = \frac{y_2-y_1}{x_2-x_1}$

### What this slope formula does is,Lets assume that a person joined a company his,

Experience (x1) = 0 
Salary     (y1) = 30000

After one year 

Experience (x2) = 1
Salary     (y2) = 40000

So How Much increment he got for one year ? 10000.

   
    slope = (40000-30000) / (1-0)
         =  10000/1 
         =  10,000
         
 Slope is giving the increment value ie, How much increment he got in one year
 


Lets take the data in a slightly different way

Experience (x1) = 1
Salary     (y1) = 40000

After 10 Years,

Experience (x2) = 10
Salary     (y2) = 150000


    Slope = (150000-40000) / (10-1)
          = 110000/9
          = 12,222.23

What is this 12,222.23 representing. It represents the average hike for each year

## ie, Slope is the change in y value with respect to one unit change in x value


so, here m means what is the average hike the company is giving to employees every year

At First we are selecting a random m and c value without any strategy. With this m and c we are creating the line. by substituting the m and c value in y = mx+c ( for each x values(YearExperience (Independent variable)) 

Example : 
       
       if the m and c values are takes as 9000 and 500 then,
       
         y' = 9000*x + 500
         
         substituting each value of x(YearExperience)
         
         y1' = 9000*1.1 + 500     (x =1.1) 
         
             = 10,400
         
         y2' = 9000*1.1 + 500     (x =1.3)
         
         This our prediction with random m and c value

         like that we are calculating the points and drawing the line
         
         
      The actual salary of person with 1.1 Year Of Experience(y1) is 39343.0(From the data) and the predicted value(y1') is 10,400
      
      Now we do the deviation calculation (y-y')^2
      
   $D_1 = (y_1-y_1')^2 = (39343-10,400)^2$
   
   Like this we calculate every deviation. In the above data we have 30 data points and we find deviation of 30 data points(D1-D30) and find the error.
   
   $ e = \frac{\sum_{}{} (y-y')^2}{N}$ 
                      
  

Then update the m and c value to some random value , and finds the error for that particular line. Repeats this steps again and again

At last, which m and c value has the least error which will be the best line of fit and we use that m and c value

### How do we change these m and c values ?

If we randomly go and update the m and c value it will take may be years to identify the best fit line. For that we are using the $ Gradient Descent Approach $
